**10. Functions**

หลังประกาศ function ให้ประกาศ :

```python
def power(x, y):
    return x**y

power(2,3)
```

```python
def summation(L):
    z = ""
    for item in L:
        z = z + item
    return z

c = summation ( ["A","B","C","D","E"] )
c
```

```python
def multi( start ,end):
    for i in range(start, end+1):
        for j in range(12):
            print(str(i) + "x" + str(j+1) + " = " + str(i*(j+1)))
        print("\n")
multi(3, 5)
```

```python
def concat(A, B):
    A = A + " First"
    B = B + " Second"
    C = A + " " + B + " Combined"
    return A,B,C

x, y, z = concat("Pladook","Here")
[x] + [y] + [z]
```
